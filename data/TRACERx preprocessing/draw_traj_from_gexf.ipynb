```python
import glob
import networkx as nx
```

```python
folder = 'path/to/trajectories_gexf/'
```

```python
for file in glob.glob(folder + '/*.gexf'):
    print(file)
    graph = nx.read_gexf(file)
    graph = nx.transitive_reduction(graph)
    dot = nx.drawing.nx_pydot.to_pydot(graph)
    out = file[:-5]
    dot.write_png(out + '.png')
```
